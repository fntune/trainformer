```python
import random
import pandas as pd
```

```python
df = pd.read_csv("s3://aisle3-ml-datasets/arcface/shopee/main.csv")
df
```

```python
clusters = df.label.unique()
clusters
```

```python
import random
test_clusters = random.choices(clusters, k=int(0.1 * len(clusters)))
len(test_clusters)
```

```python
test_df = df[df.label.isin(set(test_clusters))].reset_index(drop=True)
print(len(test_df), test_df.label.nunique())
train_df = df[~df.label.isin(set(test_clusters))].reset_index(drop=True)
print(len(train_df), train_df.label.nunique())
```

```python
test_df.to_csv("s3://aisle3-ml-datasets/arcface/shopee/test.csv")
train_df.to_csv("s3://aisle3-ml-datasets/arcface/shopee/train.csv")
```
