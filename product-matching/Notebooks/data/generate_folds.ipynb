```python
import os
from pathlib import Path

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, StratifiedKFold
```

```python
os.chdir("../../")
data_dir = Path("/data/a3-data/")
```

```python
df = pd.read_parquet(data_dir / "matched.parquet")
df
```

```python
df.columns
```

```python
df.groupby("label").image.count().sort_values(ascending=False).head(20)
```

```python
train_labels, val_labels = train_test_split(df.label.unique(), test_size=0.2)
len(train_labels), len(val_labels)
```

```python
max(train_labels), max(val_labels)
```

```python
train_df = df.loc[df.label.isin(train_labels)].reset_index(drop=True)
val_df = df.loc[df.label.isin(val_labels)].reset_index(drop=True)
train_df.to_parquet(data_dir / "train.parquet")
val_df.to_parquet(data_dir / "val.parquet")
```
